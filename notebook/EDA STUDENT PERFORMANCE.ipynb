## 3.8 Adding columns for "Total Score" and "Average"

```python
df.head()
```

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks un Reading: {reading_less_20}')
```

### 4.4.1 Gender Column

- What is the distribution of gender like?

#### Univariate Analysis (What is the distribution of gender like?)

```python
#This code uses Matplotlib and Seaborn libraries 

# Set overall style for the plots
sns.set_style("whitegrid")

# Create subplots
f,ax=plt.subplots(1,2,figsize=(10,7))

# Bar Plot
sns.countplot(
    x=df['gender'],
    data=df,
    palette =['#377eb8', '#ff7f00'], #Consistent colour scheme
    ax=ax[0],
    saturation=0.95
)

ax[0].set_title('Gender Distribution (Bar Plot)', fontsize=22) # Add title
ax[0].set_xlabel('Gender', fontsize=18) #Add x-axis label
ax[0].set_ylabel('Count', fontsize=18)  # Add y-axis label
ax[0].grid(True, linestyle='--', alpha=0.7)  # Add gridlines
ax[0].tick_params(axis='both', which='major', labelsize=14)  # Adjust tick sizes

# Label bars
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=16) # Adjust font size

# Pie Chart with custom labels
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({val:d})'
    return my_format


plt.pie(
    x=df['gender'].value_counts(),
    labels=['Male','Female'],
    explode=[0,0.1],
    autopct=autopct_format(df['gender'].value_counts()),  # Custom label format
    shadow=True,
    colors=['#377eb8', '#ff7f00']
    )

plt.title('Gender Distribution (Pie Chart)', fontsize=22) # Add title


# Add legend to pie chart
plt.legend(['Male', 'Female'], loc='upper right', fontsize=12) #Add title

#Adjust layout for spacing
plt.tight_layout(pad=3.0)

#Display the plots
plt.show()
```

#### Insights:
- There are more number of males (51.8%) than females (48.2%).

### Bivariate Analysis (Does the gender have any impact on students' performance?)

```python
df.head()
```

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
print(gender_group.index)
print(gender_group.columns)
```

```python
plt.figure(figsize=(10, 8))

#Set figure size
plt.figure(figsize=(10, 8))

# Categories for the x-axis
X = ['Total Average','Math Average']

# Extracting score for each gender from the grouped DataFrame
female_scores = [gender_group['average'][0], gender_group['math_score'][0]]
male_scores = [gender_group['average'][1], gender_group['math_score'][1]]

# Define the positions of the bars on the x-axis
X_axis = np.arange(len(X))

# Plot bars for male and female scores
plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='skyblue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='salmon')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
